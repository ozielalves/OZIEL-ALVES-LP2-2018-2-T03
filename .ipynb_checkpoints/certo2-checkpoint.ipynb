#### Importação de bibliotecas

```python
import pandas as pd
import numpy as np
import math
```

#### Leitura do arquivo csv

```python
data =  pd.read_csv('eleicao.csv', sep=';')
data.head()
```

#### Variaveis predefinidas

```python
total_vagas = 29
QE = 12684
```

#### Filtragem de coligações

```python
partidos = data['Partido/Coligação']
coligacoes = []

for i in partidos:
    elemento = i.split('-')
    if len(elemento) == 2:
        coligacoes.append(elemento[1][1:])
    else:
        coligacoes.append(elemento[0])
        
data['coligacoes'] = coligacoes
aa = data[data['coligacoes'] == 'PSD / PTB / PSL']
aa.sort_values(by='Votos', ascending=False)
```

#### Quantidade de votos de cada partido

```python
votos = data['Votos']

partidos_e_votos = {}

for i, j in enumerate(coligacoes):
    if j in partidos_e_votos :
        partidos_e_votos [j] += votos[i]
    else:
        partidos_e_votos [j] = votos[i]
```

#### Calculo do QP

```python
QP = {key: math.floor(value / QE) for key, value in partidos_e_votos.items() 
      if math.floor(value / QE) > 0}
```

#### Calculo das vagas ocupadas e restantes

```python
vagas_ocupadas =  sum(QP.values())
vagas_restantes = total_vagas - vagas_ocupadas
print(vagas_ocupadas,vagas_restantes)
```

#### Distribuição das vagas restantes

```python
for i in range(vagas_restantes):
    maior = {key : partidos_e_votos[key] / (value+1) for key, value in QP.items()}
    partido_vencedor = max(maior, key=maior.get)
    QP[partido_vencedor] += 1
```

#### Ranking de candidatos

```python
frames = []

for key, value in QP.items():
    partido_atual = data[data['coligacoes'] == key]
    partido_atual_ordenado = partido_atual.sort_values(by='Votos', ascending=False)
    frames.append(partido_atual_ordenado.iloc[0:value])
    
result = pd.concat(frames)

result_ordenado = result.sort_values(by='Votos', ascending=False).drop('coligacoes', axis=1)
```

#### Impressão de aquivo .tsv

```python
result_ordenado.to_csv('resultadoEleicoes2.tsv', sep='\t', encoding='utf-8', index = False, header=False)
```
